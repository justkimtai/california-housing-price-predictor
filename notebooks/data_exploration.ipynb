```python
import sys
import os
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_preprocessing import load_data

# Load data
df = load_data()

# Visualizing median house values distribution
sns.histplot(df["MedHouseVal"], bins=50)
plt.title("Distrubution of Median House Values")
plt.show()

# Visualizing geographical distribution
plt.scatter(df['Latitude'], df['Longitude'], c=df['MedHouseVal'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geographical Distribution of Housing Values')
plt.colorbar(label='Median House Value')
plt.show()
```
